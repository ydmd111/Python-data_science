# Importing the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the csv file:

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

# To check the salary column with sales column:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales');
```

# Also checking same columns colored according to the level of education column:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',hue='level of education');
```

# If we want to put the color option:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',hue='level of education',palette='Set2');
```

# Another color name it's possible to be used:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',hue='level of education',palette='Dark2');
```

# It's possible to devide it according to (work experience) column showed by circle size :

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',size='work experience');
```

# Another way to devide according to the circle size by  using (salary) column:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',size='salary');
```

# Also it's possible to size the circles:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',s=200);
```

# It's possible also to give density to the circles:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',s=200,alpha=0.2);
```

# Also it's possible to make multi shapesto the data , not just a circles:

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='salary',y='sales',s=200,style='level of education',hue='level of education');
```

# It's possible to use a rugplot:

```python
sns.rugplot(x='salary',data=df)
```

# The diagram above shows the data so tiny, to solve that:

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

# Also it's better to use displot:

```python
sns.displot(x='salary',data=df)
```

# It's possible to mention the number of bins or bars:

```python
sns.displot(x='salary',data=df,bins=10)
```

# It's possible to add dark grid to the diagram:

```python
sns.set(style='darkgrid')
sns.displot(x='salary',data=df,bins=10)
```

# It's possible to add white grid to the diagram:

```python
sns.set(style='whitegrid')
sns.displot(x='salary',data=df,bins=10)
```

# It's possible to add ticks among the bars in the diagram:

```python
sns.set(style='ticks')
sns.displot(x='salary',data=df,bins=10)
```

# And make the background dark:

```python
sns.set(style='dark')
sns.displot(x='salary',data=df,bins=10)
```

# And it's possible to add some additional properties to the diagram:

```python
sns.displot(x='salary',data=df,bins=10,color='red',edgecolor='blue',linewidth=4,ls='--',kde=True,rug=True);
```

# It's possible to use histplot, it's same as displot, better to use displot:

```python
sns.histplot(x='salary',data=df)
```

# And kdeplot also is possible to be used:

```python
sns.kdeplot(x='salary',data=df)
```

# In kdeplot it's to use shade:

```python
sns.kdeplot(x='salary',data=df,shade=True)
```

# It's possible to use value counts to get the total of each unique value:

```python
df['division'].value_counts()
```

# For visually counts it's better to use countplot:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='division',data=df)
```

# It's better to limit the y-axis according to minimum value:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='division',data=df)
plt.ylim(90,260)
```

# Here we have to be aware of using ylim , sometimes it decieve us:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='division',data=df)
plt.ylim(0,1000)
```

# In countplot it's possible to use hue:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',hue='division',data=df)
```

# Also it's possible to use barplot with addition of ci and estimator for example:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(x='level of education',hue='division',data=df,y='salary',estimator=np.mean,ci='sd')
plt.legend(bbox_to_anchor=(1.05,0.5))
```
