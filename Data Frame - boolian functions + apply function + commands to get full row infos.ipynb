# Using def function to create a function:

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(123456789)
```

# Another function to give a sign:

```python
def spent(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

# check it whether it's working or not:

```python
spent(10)
```

```python
spent(30)
```

# then we create another column by applying the spent function on (total_bill) column:

```python
df['spent']=df['total_bill'].apply(spent)
```

```python
df['spent']
```

# Lambda function can be used in apply function:

```python
df['total_bill'].apply(lambda num:num*2)
```

# Creating function for checking the spend and tip ratio:

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'generous'
    else:
        return 'other'
```

# Verify the function works:

```python
quality(16,1)
```

```python
quality(12,6)
```

# Applying the quality function with the use of lambda:

```python
df['quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

```python
df['quality']
```

# Or using np.vectorize with quality function:

```python
df['quality']=np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df['quality']
```
