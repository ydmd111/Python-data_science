For importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

For reading the csv file

```python
df=pd.read_csv('mpg.csv')
```

For reading the first five rows with their columns

```python
df.head()
```

For displaying the summary information in that csv file

```python
df.info()
```

For getting the description to the file

```python
df.describe().transpose()
```

For plotting the counts of (model_year) column

```python
plt.figure(figsize=(8,6),dpi=150)
sns.countplot(x=df['model_year'],data=df,palette='Set2')
```

If we want to plot the counts according to (model_year),depending on (cylinder) column

```python
plt.figure(figsize=(8,4),dpi=150)
sns.countplot(x=df['model_year'],data=df,palette='Set1',hue='cylinders')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

To get the unique data of (model_year) column

```python
df['model_year'].unique()
```

To count each value in (model_year) column

```python
df['model_year'].value_counts()
```

When we getting the mean or the average of the file by indexing (model_year) and (cylinder) columns

```python
df.groupby(['model_year','cylinders']).mean()
```

For getting the the column names when the multi-index is there 

```python
df.groupby(['model_year','cylinders']).mean().columns
```

To get the the index

```python
df.groupby(['model_year','cylinders']).mean().index
```

To get a single column data

```python
df.groupby(['model_year','cylinders']).mean()['mpg']
```

To get the description when (model_year) is the index and transpose the rows and columns

```python
df.groupby('model_year').describe().transpose()
```

To save the the groupby mean function when (model_year) and (cylinders)columns are index

```python
year_cyl=df.groupby(['model_year','cylinders']).mean()
year_cyl
```

To get year_cyl's index names

```python
year_cyl.index.names
```

To get thindex data or index levels

```python
year_cyl.index.levels
```

If we want to collect info according to (model_year) index of a single year only

```python
year_cyl.loc[70]
```

If we want to collect info according to (model_year) index of a multiple year 

```python
year_cyl.loc[[70,82]]
```

If we want to get information according to both indexes

```python
year_cyl.loc[(70,4)]
```

Also we can get information by mentioning the key and level

```python
year_cyl.xs(key=70,level='model_year')
```

Also we can use another level

```python
year_cyl.xs(key=4,level='cylinders')
```

If we want to get a full information of rows according to special condition

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

If we need to swap the index accordingly

```python
year_cyl.swaplevel()
```

If we want to sort date descendingly according to (model_year) column

```python
year_cyl.sort_index(level='model_year',ascending=False)
```

If we want to apply a multi functions

```python
df.agg(['std','mean'])
```

If we want to apply multi function on a single column

```python
df.agg(['std','mean'])['mpg']
```

If we want to apply functions on multi-columns ,on each column applies multi and different functions

```python
df.agg({'mpg':['max','mean'],'weight':['mean','std']})
```
