# Importing pandas and numpy libraries:

```python
import numpy as np
import pandas as pd
```

# Creating x value:

```python
x=np.linspace(0,11,10)
```

```python
x
```

# For plotting , it's needed to import matplotlib lirary:

```python
import matplotlib.pyplot as plt
```

# Now for the ploting:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x)
ax.plot(x,x**2)
```

# If we need to add a legend:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend();
```

# If we want to mention the legend location:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend(loc='lower left');
```

# Also:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend(loc=4);
```

# It's possible to locate the legend outside the diagram:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend(loc=(1.1,0.5));
```

# Also:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend(loc=(0.5,0.5));
```

# It's possible also to locate the legend  in the left side of the diagram:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')
ax.legend(loc=(-0.5,0.5));
```

# To drow 2 straight lines with mentioning the line colors:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='orange',label='x vs x')
ax.plot(x,x+1,color='blue',label='x vs x+1')
ax.legend()
```

# To add the line width:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='orange',lw=10)
ax.plot(x,x+1,color='blue',linewidth=5)
```

# Also it's possible to add line styles:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='orange',lw=10,ls='--')
ax.plot(x,x+1,color='blue',linewidth=5,linestyle='-.')
```

# It's possible to add line style by using numbers:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lines=ax.plot(x,x+1,color='blue',linewidth=5)
lines[0].set_dashes([5,2,10,2])
```

# Also

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lines=ax.plot(x,x+1,color='blue',linewidth=5)
lines[0].set_dashes([1,2,1,2,10,2])
```

# To add marker to the line:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x+1,color='blue',linewidth=2,marker='o')
```

# To add line style with marker:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x+1,color='blue',linewidth=2,marker='o',ls='-.')
```

# There are many markers can be used:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='red',marker='+',ls='-.')
ax.plot(x,x+1,color='blue',marker='o',linestyle='--')
```

# It's possible to add the marker size also:

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x+1,color='blue',marker='o',linestyle='--',markersize=20)
```

# Again it's possible to add some styles to the 

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x+1,color='blue',marker='o',linestyle='--',markersize=20,markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange')
```
