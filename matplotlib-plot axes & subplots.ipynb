# Importing the needed libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Creating and b values:

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
a
```

```python
b
```

# Now x and y:

```python
x=np.arange(0,10)
y=2*x
```

```python
x
```

```python
y
```

# Create multiple axes, small and large together:

```python
fig=plt.figure()
# Large Axes
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('Power of 4')
# Small Axes
axes2=fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('Zoom In')
plt.show()
```

# Now let's create a subplot axes,fig:

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

# Now to create 2 axes row and 2 axes columns with plot:

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

# Now figure out new axes , 1 row & 3 columns:

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
axes[0].plot(x,y)
```

# Making for loop to fill the axes:

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
for ax in axes:    
    ax.plot(x,y)
```

# To fill the first axes with a proper distance between each axes:

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
axes[0].plot(x,y)
fig.tight_layout()
```

# Create 4 axes and adding x & y labels and title to each axes: 

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel('X ax 0,0')
axes[0][0].set_ylabel('y ax 0,0')
axes[0][0].set_title('x,y')
axes[0][1].plot(x,y)
axes[0][1].set_xlabel('X ax 0,1')
axes[0][1].set_ylabel('y ax 0,1')
axes[0][1].set_title('x,y')
axes[1][0].plot(x,y)
axes[1][0].set_xlabel('X ax 1,0')
axes[1][0].set_ylabel('y ax 1,0')
axes[1][0].set_title('x,y')
axes[1][1].plot(a,b)
axes[1][1].set_xlabel('X ax 1,1')
axes[1][1].set_ylabel('y ax 1,1')
axes[1][1].set_title('a,b')
fig.tight_layout()
```
