# To import the needed libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# To read and display the csv file:

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

# To use boxplot , if y first then the diagram will be up and down:

```python
sns.boxplot(data=df,y='math score')
```

# Y is first and x second:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='test preparation course')
```

# It's possible to add hue to boxplot:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# If x comes first then the diagram will be rotated 90 degrees:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,x='math score',y='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# To use violinplot:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education')
```

# Adding hue and the violin devided into 2 halves:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True)
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# If needed to remove the white spot in the middle which represents the median value:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True,inner=None)
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# If we need to devide the diagram into quarters:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True,inner='quartile')
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# If we need to add lines to the diagrams:

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',inner='stick')
```

# If we want to add bandwidth to the diagram:

```python
sns.violinplot(data=df,x='reading score',y='parental level of education',bw=0.1)
```

# Minimising the bandwidth:

```python
sns.violinplot(data=df,x='reading score',y='parental level of education',bw=0.01)
```

# Making the bandwidth higher:

```python
sns.violinplot(data=df,x='reading score',y='parental level of education',bw=0.5)
```

# Again higher:

```python
sns.violinplot(data=df,x='reading score',y='parental level of education',bw=1)
```

# To use swarmplot:

```python
sns.swarmplot(data=df,x='math score',y='gender',size=2)
```

# To add hue to the diagram:

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course')
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# There the hue is mixed up, to solve that problem:

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# To use boxenplot:

```python
sns.boxenplot(data=df,x='math score',y='test preparation course',hue='gender')
plt.legend(bbox_to_anchor=(1.25,0.5))
```

# Also pairplot is useful to be used:

```python
sns.pairplot(data=df)
```

# It's possible to add hue to pairplot ,automatically the displot will be converted into kdeplot:

```python
sns.pairplot(data=df,hue='gender')
```

# It's possible to remove the repeated diagrams in pairplot:

```python
sns.pairplot(data=df,hue='gender',corner=True)
```

# Using PairPlot:

```python
sns.PairGrid(df)
```

# No diagram has been added, to solve that problem we should add it manually:

```python
g=sns.PairGrid(df,hue='gender')
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot,shade=True)
g=g.map_lower(sns.kdeplot)
g=g.add_legend()
```
